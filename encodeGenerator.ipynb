```python
import cv2
import face_recognition
import os
import firebase_admin
from firebase_admin import credentials, initialize_app, storage
import pyrebase
import pickle
import numpy as np

# Check if Firebase app is already initialized
try:
    firebase_admin.get_app(name='my_app')
except ValueError as e:
    # If Firebase app is not initialized, initialize it with a unique name
    cred = credentials.Certificate("C:\\Users\\user\\Downloads\\votingService key.json")
    initialize_app(cred, {
        "databaseURL": "https://votingfacerecognition-default-rtdb.firebaseio.com/",
        "storageBucket": 'votingfacerecognition.appspot.com'
    }, name='my_app')

# Get the storage bucket for the app
bucket = storage.bucket(app=firebase_admin.get_app(name='my_app'))

# Importing student images
folderModePath = "images"
PathList = os.listdir(folderModePath) # list of files in folder

imgList = []
studentId = []

for path in PathList:
    path = path.strip()
    if not path:
        continue

    imgList.append(cv2.imread(os.path.join(folderModePath, path)))
    studentId.append(os.path.splitext(path)[0])

    fileName = f'{folderModePath}/{path}'         # create folder----> images
    blob = bucket.blob(path)
    blob.upload_from_filename(fileName)
   
print("student id:", studentId)
print("Number of loaded images:", len(imgList))

# Load existing encodings from file
file = open("encodeFile.p", 'rb')
encodeListKnownWithIds = pickle.load(file)
file.close()
encodeListKnown, existingIds = encodeListKnownWithIds

def findEncodings(imgList):
    encodeList = []
    for img in imgList:
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        face_locations = face_recognition.face_locations(img_rgb)
        face_encodings = face_recognition.face_encodings(img_rgb, face_locations)
        for encode in face_encodings:
            # Check if the encoding already exists
            exists = False
            for known_encode in encodeListKnown:
                if np.array_equal(encode, known_encode):
                    print("The image is already present.")
                    exists = True
                    break
            if not exists:
                encodeList.append(encode)
    return encodeList


print("Encoding started")
encodeListNew = findEncodings(imgList)
print("Encoding complete")

# Append new encodings to the existing list
encodeListKnown += encodeListNew
studentId += [id for id in existingIds if id not in studentId]

# Save updated encodings to file
encodeListKnownWithIds = [encodeListKnown, studentId]
file = open("encodeFile.p", 'wb')
pickle.dump(encodeListKnownWithIds, file)
file.close()
print("File saved")
```
