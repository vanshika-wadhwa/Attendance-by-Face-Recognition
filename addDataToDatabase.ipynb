```python
import firebase_admin
from firebase_admin import credentials
from firebase_admin import db
import sqlite3

def initialize_firebase_app():
    # Check if Firebase app is already initialized
    if not firebase_admin._apps:
        # If not initialized, initialize the app
        cred = credentials.Certificate("C:\\Users\\user\\Downloads\\votingService key.json")
        firebase_admin.initialize_app(cred, {
            "databaseURL": "https://votingfacerecognition-default-rtdb.firebaseio.com/"
        })
        print("Firebase app initialized successfully.")
    else:
        print("Firebase app already initialized.")

def connect_to_sqlite_database(database_name):
    # Connect to SQLite database
    conn = sqlite3.connect(database_name)
    cursor = conn.cursor()
    print("Connected to SQLite database:", database_name)
    return conn, cursor

def create_sqlite_table_if_not_exists(cursor):
    # Create SQLite table if not exists
    create_table_sql = '''
        CREATE TABLE IF NOT EXISTS firebase_data (
            key TEXT PRIMARY KEY,
            name TEXT,
            major TEXT,
            starting_year INTEGER,
            total_attendance INTEGER,
            standing TEXT,
            year INTEGER,
            last_attendance_time TEXT
        )
    '''
    cursor.execute(create_table_sql)
    print("SQLite table 'firebase_data' created successfully.")

def insert_data_into_sqlite(conn, cursor, data):
    # Function to insert data into SQLite
    for key, value in data.items():
        cursor.execute("INSERT OR REPLACE INTO firebase_data VALUES (?, ?, ?, ?, ?, ?, ?, ?)", (
            key,
            value.get("name", ""),
            value.get("major", ""),
            value.get("starting_year", 0),
            value.get("total_attendance", 0),
            value.get("standing", ""),
            value.get("year", 0),
            value.get("last_attendance_time", "")
        ))
    conn.commit()
    print("Data inserted into SQLite successfully.")

def main():
    # Initialize Firebase app
    initialize_firebase_app()

    # Connect to SQLite database
    conn, cursor = connect_to_sqlite_database('realtimedatabaseSql.db')

    # Create SQLite table if not exists
    create_sqlite_table_if_not_exists(cursor)

    # Get a reference to the Firebase Realtime Database
    ref = db.reference('students')

    # Fetch data from Firebase
    firebase_data = ref.get()

    # Insert data into SQLite
    if firebase_data:
        insert_data_into_sqlite(conn, cursor, firebase_data)

    # Close SQLite connection
    conn.close()
    print("SQLite connection closed.")

if __name__ == "__main__":
    main()
```
